# Section 3

```python
import numpy as np
class Perceptron:
    def __init__(self, learning_rate=0.1, epochs=10):
        self.lr = learning_rate
        self.epochs = epochs

    def fit(self, X, y):
        self.weights = np.zeros(X.shape[1])
        self.bias = 0

        for _ in range(self.epochs):
            for i in range(len(X)):
                y_pred = self.predict(X[i])
                error = y[i] - y_pred
                self.weights += self.lr * error * X[i]
                self.bias += self.lr * error

    def predict(self, x):
        linear_output = np.dot(x, self.weights) + self.bias
        return 1 if linear_output >= 0 else 0
```

```python
# AND gate data
X_and = np.array([[0,0], [0,1], [1,0], [1,1]])
y_and = np.array([0, 0, 0, 1])

# Train perceptron
and_perceptron = Perceptron()
and_perceptron.fit(X_and, y_and)

# Test AND gate
print("AND Gate Results:")
for x in X_and:
    print(f"{x} -> {and_perceptron.predict(x)}")
```

```python
# OR gate data
X_or = np.array([[0,0], [0,1], [1,0], [1,1]])
y_or = np.array([0, 1, 1, 1])

# Train perceptron
or_perceptron = Perceptron()
or_perceptron.fit(X_or, y_or)

# Test OR gate
print("\nOR Gate Results:")
for x in X_or:
    print(f"{x} -> {or_perceptron.predict(x)}")
```

# Section 4

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
# XOR input and output
X = torch.tensor([[0., 0.],
                  [0., 1.],
                  [1., 0.],
                  [1., 1.]])

y = torch.tensor([[0.],
                  [1.],
                  [1.],
                  [0.]])
```

```python
class XORNet(nn.Module):
    def __init__(self):
        super(XORNet, self).__init__()
        self.hidden = nn.Linear(2, 4)
        self.output = nn.Linear(4, 1)

    def forward(self, x):
        x = torch.relu(self.hidden(x))
        x = torch.sigmoid(self.output(x))
        return x
```

```python
model = XORNet()

criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)
```

```python
epochs = 5000

for epoch in range(epochs):
    optimizer.zero_grad()
    output = model(X)
    loss = criterion(output, y)
    loss.backward()
    optimizer.step()

    if epoch % 1000 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item():.4f}")
```

```python
with torch.no_grad():
    predictions = model(X)
    predictions = (predictions > 0.5).float()

print("\nXOR Predictions:")
for i in range(len(X)):
    print(f"{X[i].tolist()} -> {predictions[i].item()}")
```

# Section 5

```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
```

```python
# Set seed for reproducibility
torch.manual_seed(42)

# Generate input data
X = torch.rand(100, 2)

# Regression target
y = 3 * X[:, 0:1] + 2 * X[:, 1:2]
```

```python
class SimpleRegressionNN(nn.Module):
    def __init__(self):
        super(SimpleRegressionNN, self).__init__()

        # Input → Hidden (2 neurons)
        self.hidden = nn.Linear(2, 2)

        # Hidden → Output (1 neuron)
        self.output = nn.Linear(2, 1)

    def forward(self, x):
        h = torch.relu(self.hidden(x))  # f(z¹)
        out = self.output(h)            # Linear output f(z²)
        return out
```

```python
model = SimpleRegressionNN()

criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)
```

```python
epochs = 1000
losses = []

for epoch in range(epochs):
    optimizer.zero_grad()

    predictions = model(X)
    loss = criterion(predictions, y)

    loss.backward()
    optimizer.step()

    losses.append(loss.item())

    if epoch % 200 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item():.6f}")
```

```python
with torch.no_grad():
    test_sample = torch.tensor([[0.5, 0.2]])
    pred = model(test_sample)

print("Predicted Output:", pred.item())
print("Expected Output:", 3*0.5 + 2*0.2)
```
